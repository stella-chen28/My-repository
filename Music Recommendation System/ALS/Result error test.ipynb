```python
rlt = pd.read_csv('result_oriID.csv')
dic_rlt = rlt.set_index('ID')
dic_rlt[1:5]
```

```python
dic_rlt = dic_rlt.T.to_dict('list')
test_set=pd.read_csv("test_set_oriID.csv")
```

```python
test_set['w'] = test_set.target+1
test_set[0:5]
```

```python
test_set = test_set.loc[test_set.target==1]
test_set[0:5]
```

```python
no_rec={'no_rec_user':[]}
N = 0
nor = 0
for i in range(test_set.shape[0]):
    n = 0
    rs = dic_rlt.get(test_set.iloc[i,0])
    if rs !=None:
        for k in range(10):
            if test_set.iloc[i,1]==rs[k]:
                N=N+1
                break
    else:
        no_rec['no_rec_user'].append(test_set.iloc[i,0])
        nor = nor+1
        print("miss recommend:")
        print(nor)
    if i%1000==0:
        print(i)
        print(N)
        print('---')
```

```python
N
```

```python
1-N/((len(test_set.msno.drop_duplicates())-nor)*10)
```
