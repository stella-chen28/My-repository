```python
import pandas as pd
```

```python
data_1 = pd.read_csv('/home/jb4076/music/1.csv',converters={i: str for i in range(0, 100)})
data_2 = pd.read_csv('/home/jb4076/music/2.csv',converters={i: str for i in range(0, 100)})
data_3 = pd.read_csv('/home/jb4076/music/3.csv',converters={i: str for i in range(0, 100)})
data_4 = pd.read_csv('/home/jb4076/music/4.csv',converters={i: str for i in range(0, 100)})
data_6 = pd.read_csv('/home/jb4076/music/6.csv',converters={i: str for i in range(0, 100)})
data_7 = pd.read_csv('/home/jb4076/music/7.csv')
data_9 = pd.read_csv('/home/jb4076/music/genre_count_feature.csv')
data_10 = pd.read_csv('/home/jb4076/music/song_count_feature.csv')
```

```python
print(data_1.head())
print(data_2.head())
print(data_3.head())
data_4.columns = ['id','msno','lang_repeat_top1','lang_repeat_top2','lang_repeat_top3']
print(data_4.head())
print(data_6.head())
print(data_7.head())
print(data_9.head())
print(data_10.head())
```

```python
data_merge_1 = data_1.merge(data_2[['msno','genre_repeat_top1','genre_repeat_top2','genre_repeat_top3',
                    'genre_repeat_top4','genre_repeat_top5','genre_repeat_top6',
                    'genre_repeat_top7','genre_repeat_top8']], on = 'msno')
```

```python
data_merge_1.head()
```

```python
data_merge = data_merge_1.merge(data_3[['msno','lang1','lang2','lang3']], on = 'msno')\
                         .merge(data_4[['msno','lang_repeat_top1','lang_repeat_top2',
                                       'lang_repeat_top3']], on='msno')\
                         .merge(data_6[['msno','artist_name1','artist_name2',
                                       'artist_name3']], on = 'msno')\
                         .merge(data_7[['msno','mean_song_len']],on='msno')\
                         .merge(data_9[['msno','genre_listen_num','genre_repeat_num']], on='msno')\
                         .merge(data_10[['msno','song_listen_num','song_repeat_num']], on='msno')
```

```python
pd.set_option('display.max_columns', 100)
display(data_merge)
```

```python
data_merge.to_csv('./feature_data_total_1.csv', encoding='utf-8', index=False)
```

```python
data_numeric_feature = data_merge[['msno','bd','mean_song_len','genre_listen_num',
                                   'genre_repeat_num','song_listen_num','song_repeat_num']]
```

```python
data_numeric_feature.to_csv('./feature_data_numeric_1.csv', encoding='utf-8', index=False)
```
