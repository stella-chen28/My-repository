```python
train = pd.read_csv('/Users/minta/Desktop/5291/project/training.csv')
members = pd.read_csv('/Users/minta/Desktop/5291/project/members.csv')
```

```python
print(train.head())
```

```python
train['target_num'] = train['target']+1
print(train)
```

```python
#data groupby 
group = train.groupby(["msno","artist_name"])['target_num'].sum()
print(group)
type(group)
```

```python
#series to dataframe
group1=group.to_frame(name='target_num').reset_index()
type(group1)
group1
```

```python
art_df=group1.sort_values(['msno',"target_num"],ascending=False).groupby('msno').head(3)
art_df
```

```python
#set a new dataframe
df = pd.DataFrame(columns= ['msno','artist_name1','artist_name2','artist_name3'])
#find unique msno
unique_msno=art_df['msno'].unique()

i = 0
for msno in unique_msno:
#find the start position of every unique msno
    index_list = art_df.loc[art_df['msno'] == msno].index.tolist()
    df.at[i,'msno'] = msno
#find  the Top3 art_name an
    if len(index_list) >= 3 :
        df.at[i,'artist_name1'] = art_df.loc[index_list[0]]['artist_name']
        df.at[i,'artist_name2'] = art_df.loc[index_list[1]]['artist_name']
        df.at[i,'artist_name3'] = art_df.loc[index_list[2]]['artist_name']
    elif len(index_list) == 2 :
        df.at[i,'artist_name1'] = art_df.loc[index_list[0]]['artist_name']
        df.at[i,'artist_name2'] = art_df.loc[index_list[1]]['artist_name']
    else:
        df.at[i,'artist_name1'] = art_df.loc[index_list[0]]['artist_name']
    i += 1
print(df.head())
```

```python
#data merge and save to csv
members_update = members[(members.msno.isin(train.msno))]
members_merge = members_update.merge(df, on = 'msno')
members_merge.to_csv('./members_add_artist.csv', encoding='utf-8', index=False)
```
