```python
import seaborn as sns
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt

sns.set(style="ticks", color_codes=True)
```

```python
members_genre = pd.read_csv('/Users/jbian/Desktop/CU/5291/project/members_add_genre.csv')
```

```python
genre_count_1 = Counter(members_genre.genre_1)
genre_count_2 = Counter(members_genre.genre_2)
genre_count_3 = Counter(members_genre.genre_3)
genre_count_4 = Counter(members_genre.genre_4)
genre_count_5 = Counter(members_genre.genre_5)

x_values_1 = list(genre_count_1.keys())
y_values_1 = list(genre_count_1.values())
y_values_11 = [x/members_genre.shape[0] for x in list(genre_count_1.values())]

x_values_2 = list(genre_count_2.keys())
y_values_2 = list(genre_count_2.values())
y_values_22 = [x/members_genre.shape[0] for x in list(genre_count_2.values())]

x_values_3 = list(genre_count_3.keys())
y_values_3 = list(genre_count_3.values())
y_values_33 = [x/members_genre.shape[0] for x in list(genre_count_3.values())]

x_values_4 = list(genre_count_4.keys())
y_values_4 = list(genre_count_4.values())
y_values_44 = [x/members_genre.shape[0] for x in list(genre_count_4.values())]

x_values_5 = list(genre_count_5.keys())
y_values_5 = list(genre_count_5.values())
y_values_55 = [x/members_genre.shape[0] for x in list(genre_count_5.values())]
```

```python
fig = plt.figure(figsize=(12, 15)) 
fig = plt.subplot(5,2,1)
sns.barplot(x=x_values_1, y=np.log10(y_values_1))
fig.set_ylabel('Log10 Values of Count')
fig.set_title('Distributions of people\'s favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,2)
sns.barplot(x=x_values_1, y=y_values_11)
fig.set_ylabel('Frequency')
fig.set_title('Distributions of people\'s favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,3)
sns.barplot(x=x_values_2, y=np.log10(y_values_2))
fig.set_ylabel('Log10 Values of Count')
fig.set_title('Distributions of people\'s second favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,4)
sns.barplot(x=x_values_2, y=y_values_22)
fig.set_ylabel('Frequency')
fig.set_title('Distributions of people\'s second favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,5)
sns.barplot(x=x_values_3, y=np.log10(y_values_3))
fig.set_ylabel('Log10 Values of Count')
fig.set_title('Distributions of people\'s third favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,6)
sns.barplot(x=x_values_3, y=y_values_33)
fig.set_xlabel('Genres')
fig.set_ylabel('Frequency')
fig.set_title('Distributions of people\'s third favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,7)
sns.barplot(x=x_values_4, y=np.log10(y_values_4))
fig.set_xlabel('Genres')
fig.set_ylabel('Log10 Values of Count')
fig.set_title('Distributions of people\'s forth favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,8)
sns.barplot(x=x_values_4, y=y_values_44)
fig.set_ylabel('Frequency')
fig.set_title('Distributions of people\'s forth favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,9)
sns.barplot(x=x_values_5, y=np.log10(y_values_5))
fig.set_xlabel('Genres')
fig.set_ylabel('Log10 Values of Count')
fig.set_title('Distributions of people\'s fifth favorite genre')
fig.set_xticklabels(labels = '')

fig = plt.subplot(5,2,10)
sns.barplot(x=x_values_5, y=y_values_55)
fig.set_xlabel('Genres')
fig.set_ylabel('Frequency')
fig.set_title('Distributions of people\'s fifth favorite genre')
fig.set_xticklabels(labels = '')

# plt.savefig('./genre_dist.jpg')
```
